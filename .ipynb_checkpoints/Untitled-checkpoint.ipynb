6. (10 pts.) Compute the mean median household income for Democratic counties and Republican counties. Which one is higher? Perform a hypothesis test to determine whether this difference is statistically significant at the ùú∂=ùüé.ùüéùüì significance level. What is the result of the test? What conclusion do you make from this result?

```python
merged_demographics.loc[merged_demographics['Party'] == 1, 'Median Household Income'].head()
```

```python
median_democrat_mean = merged_demographics.loc[merged_demographics['Party'] == 1, 'Median Household Income'].mean()
```

```python
print(median_democrat_mean)
```

```python
median_republican_mean = merged_demographics.loc[merged_demographics['Party'] == 0, 'Median Household Income'].mean()
```

```python
print(median_republican_mean)
```

```python
median_dem_std = merged_demographics.loc[merged_demographics['Party'] == 1, 'Median Household Income'].std()
median_rep_std = merged_demographics.loc[merged_demographics['Party'] == 0, 'Median Household Income'].std()
print(median_dem_std)
print(median_rep_std)
```

```python
dem_arr = merged_demographics.loc[merged_demographics['Party'] == 1, 'Median Household Income']
rep_arr = merged_demographics.loc[merged_demographics['Party'] == 0, 'Median Household Income']
[statistic, pval] = st.ttest_ind(dem_arr, rep_arr, equal_var = False)
print(statistic)
print(pval)
```

```python
dem_arr.head()
```

Answer:

The Democrat median household income mean is higher than the Republican, 53766.456 v. 48724.615.

The p-value for a two-sample t-test comes to 6.536254891102229e-08, which is far below the alpha value of 0.05. This means that we reject the null hypothesis and we can see that the difference in median income between Democratic counties and Republican counties is statistically significant. 

When looking at the median household income mean between the two, it does not seem that there is such a large difference between the two. However, the p-value contradicts this assessment, and leads me to believe that there is a large difference in income level between democrat and republican counties. 



7. (10 pts.) Compute the mean population for Democratic counties and Republican counties. Which one is higher? Perform a hypothesis test to determine whether this difference is statistically significant at the ùú∂ = ùüé. ùüéùüì significance level. What is the result of the test? What conclusion do you make from this result?

```python
dem_mean_pop = merged_demographics.loc[merged_demographics['Party'] == 1, 'Total Population'].mean()
rep_mean_pop = merged_demographics.loc[merged_demographics['Party'] == 0, 'Total Population'].mean()

print(dem_mean_pop)
print(rep_mean_pop)
```

```python
dem_arr = merged_demographics.loc[merged_demographics['Party'] == 1, 'Total Population']
rep_arr = merged_demographics.loc[merged_demographics['Party'] == 0, 'Total Population']

[statistic, pval] = st.ttest_ind(dem_arr, rep_arr, equal_var = False)

print(pval)
```

Answer:
    
The Democrat total population mean is higher than the Republican, 299308.722 v. 54041.167.

The p-value for a two-sample t-test comes to 2.2795809094677384e-14, which is far below the alpha value of 0.05. This means that we reject the null hypothesis and we can see that the difference in total population between Democratic counties and Republican counties is highly statistically significant. 

Both the total population means and the p-value for the two-sample t-test shows that there is a huge difference in population between Democratic and Republican counties.
